# 亞洲

```python
import pandas as pd
import seaborn as sns
%config InlineBackend.figure_format = 'retina'

df = pd.read_csv('/Users/ziyon/Documents/Business Programming/Final Project/asia.csv')

df['Freedom and Equality'] = df['Freedom to make life choices'] + df['Perceptions of corruption']
df

fe5 = df.nlargest(5, 'Freedom and Equality')
fe5

as_fe5 = sns.barplot(x='Country',y='Freedom and Equality', data=fe5)
as_fe5.tick_params(labelsize=8)

for i in as_fe5.patches:
    height = round(i.get_height(),3)
    as_fe5.annotate(f'{height}', (i.get_x() + i.get_width() / 2, i.get_height()), ha='center', va='bottom')

as_fe5
```
