# 亞洲S+G

```python
import pandas as pd

df = pd.read_csv('/Users/ziyon/Documents/Business Programming/Final Project/asia.csv')

df
```

```python
df['Social Generosity'] = df['Social support'] + df['Generosity']
df
```

```python
sg5 = df.nlargest(5, 'Social Generosity')
sg5
```

```python
import seaborn as sns
%config InlineBackend.figure_format = 'retina'

as_sg5 = sns.barplot(x='Country',y='Social Generosity', data=sg5)

#數值標籤
for p in as_sg5.patches:
    height = round(p.get_height(),3)
    as_sg5.annotate(f'{height}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
```
